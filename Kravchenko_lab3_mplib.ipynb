```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from google.colab import files
import numpy as np
```

```python
uploaded = files.upload()
data = pd.read_excel("task1.xlsx")
data.head
```

```python
x = data.values[:, 0]
df=pd.DataFrame(x)
df = pd.to_datetime(df[0]).dt.month  #преобразовал значения дат в месяцы 
y = data.values[:, [1,2,3,4]]
plt.plot(df,y,label=["Open","High","Low","Close"])
plt.legend()
plt.xticks(df)
plt.xlabel("2016, месяцы")
plt.savefig("trinket_plot.png")
#files.download("trinket_plot.png") # скачиваю график
plt.show()
```
